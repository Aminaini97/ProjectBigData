```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# Tambahkan ini sebelum apply()
factory = StopWordRemoverFactory()
stopword_remover = factory.create_stop_word_remover()

def clean_text(text):
    import re
    if not isinstance(text, str) or not text.strip():
        return ""
    text = text.lower()
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.encode('ascii', 'ignore').decode('ascii')
    text = re.sub(r'\s+', ' ', text).strip()
    text = stopword_remover.remove(text)
    return text

df["clean_text"] = df["text"].apply(clean_text)

# Simpan hasil
df.to_csv("youtube_comments_clean.csv", index=False)
print(f"‚úÖ Cleaning selesai! Total data bersih: {len(df)}")
```

```python
!pip install deep-translator -q

from deep_translator import GoogleTranslator
import time
import pandas as pd

# Contoh: df["clean_text"] sudah ada dari hasil sebelumnya
# df = pd.read_csv("hasil_clean.csv")

translator = GoogleTranslator(source='id', target='en')

def batch_translate_safe(texts, batch_size=100):
    results = []
    total = len(texts)

    for i in range(0, total, batch_size):
        batch = texts[i:i + batch_size]
        print(f"üîÑ Menerjemahkan batch {i}-{i + len(batch)} ...")
        try:
            batch_translated = []
            for text in batch:
                if not isinstance(text, str) or text.strip() == "":
                    batch_translated.append("")
                    continue
                translated = translator.translate(text)
                batch_translated.append(translated)
                time.sleep(0.3)  # jeda antar permintaan
            results.extend(batch_translated)
        except Exception as e:
            print(f"‚ö†Ô∏è Batch gagal diterjemahkan ({i}-{i + len(batch)}): {e}")
            # Tambahkan placeholder agar panjang tetap sama
            results.extend([""] * len(batch))
            time.sleep(5)

    return results

# Jalankan terjemahan
df["translated"] = batch_translate_safe(df["clean_text"].tolist())

# Simpan hasilnya
df.to_csv("hasil_terjemahan.csv", index=False)
print("‚úÖ Translasi selesai dan disimpan ke 'hasil_terjemahan.csv'")
```

# Pembersihan data

```python
# Instal pustaka emoji
!pip install emoji

import re
import string
import pandas as pd
import emoji # Import library emoji

# Atur opsi agar pandas menampilkan semua baris dan kolom
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Muat data dari file CSV yang sudah kamu simpan
df_comments = pd.read_csv("komentar_aura_farming.csv")

def clean_text_final(text):
    """Fungsi untuk membersihkan teks secara komprehensif, termasuk emoji."""
    text = str(text)

    # 1. Hapus tautan URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 2. Hapus semua emoji menggunakan library 'emoji'
    text = emoji.demojize(text)
    text = re.sub(r':[a-zA-Z0-9_]+:', '', text)

    # 3. Ubah semua teks menjadi huruf kecil
    text = text.lower()

    # 4. Hapus tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 5. Hapus angka
    text = re.sub(r'\d+', '', text)

    # 6. Hapus spasi berlebihan dan trim
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Terapkan fungsi pembersihan yang diperbarui ke kolom 'text'
print("Melakukan pra-pemrosesan data secara komprehensif...")
df_comments['cleaned_text'] = df_comments['text'].apply(clean_text_final)

# Tampilkan semua baris data yang sudah bersih
print("\nData komentar setelah dibersihkan:")
print(df_comments[['text', 'cleaned_text']])

# Simpan DataFrame yang sudah diproses ke file baru
df_comments.to_csv("komentar_aura_farming_bersih_final.csv", index=False)
print("\nData yang sudah bersih (final) disimpan ke file: komentar_aura_farming_bersih_final.csv")
```

# Translate

```python
# Instal pustaka yang dibutuhkan
!pip install googletrans==4.0.0-rc1

import pandas as pd
from googletrans import Translator

# Muat data yang sudah dibersihkan dari langkah sebelumnya
df_comments = pd.read_csv("komentar_aura_farming_bersih_final.csv")

# Inisialisasi penerjemah
translator = Translator()

# Fungsi untuk menerjemahkan teks
def translate_text(text):
    """Menerjemahkan teks ke Bahasa Inggris, melewati teks kosong atau non-string."""
    if pd.isna(text) or not isinstance(text, str) or len(text.strip()) == 0:
        return ""
    try:
        # Menentukan bahasa sumber secara otomatis dan menerjemahkan ke Inggris
        return translator.translate(text, dest='en').text
    except Exception as e:
        print(f"Error menerjemahkan: {e}")
        return ""

# Terapkan fungsi penerjemahan ke kolom 'cleaned_text'
print("Melakukan penerjemahan komentar...")
df_comments['translated_text'] = df_comments['cleaned_text'].apply(translate_text)

# Tampilkan beberapa komentar yang sudah diterjemahkan
print("\nBeberapa komentar setelah diterjemahkan ke Bahasa Inggris:")
print(df_comments[['text', 'cleaned_text', 'translated_text']].head(10))

# Simpan data yang sudah diterjemahkan
df_comments.to_csv("komentar_aura_farming_translated.csv", index=False)
print("\nData dengan kolom terjemahan sudah disimpan.")
```

```python
# Instal pustaka yang dibutuhkan
!pip install nltk

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

# Atur opsi agar pandas menampilkan semua baris dan kolom
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Unduh kamus VADER yang diperlukan
# Ini hanya perlu dilakukan sekali
nltk.download('vader_lexicon')

# Muat data yang sudah dibersihkan dari langkah sebelumnya
df_comments = pd.read_csv("komentar_aura_farming_bersih_final.csv")

# Inisialisasi VADER Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

# Fungsi untuk mengklasifikasikan sentimen
def classify_sentiment(text):
    """
    Mengklasifikasikan sentimen berdasarkan skor VADER.
    Skor 'compound' > 0.05 dianggap Positif.
    Skor 'compound' < -0.05 dianggap Negatif.
    Selain itu, dianggap Netral.
    """
    if pd.isna(text):
        return 'Netral'

    score = analyzer.polarity_scores(text)
    compound_score = score['compound']

    if compound_score > 0.05:
        return 'Positif'
    elif compound_score < -0.05:
        return 'Negatif'
    else:
        return 'Netral'

# Terapkan fungsi ke kolom 'cleaned_text'
print("Melakukan analisis sentimen pada data yang sudah bersih...")
df_comments['sentiment'] = df_comments['cleaned_text'].apply(classify_sentiment)

# Tampilkan semua baris data dengan label sentimen
# Perhatikan, `.head(10)` sudah dihapus
print("\nData komentar dengan label sentimen:")
print(df_comments[['text', 'cleaned_text', 'sentiment']])

# Hitung jumlah total sentimen
sentiment_counts = df_comments['sentiment'].value_counts()
print("\nRingkasan total sentimen:")
print(sentiment_counts)

# Simpan DataFrame yang sudah ada kolom sentimennya
df_comments.to_csv("komentar_aura_farming_sentimen.csv", index=False)
print("\nData dengan sentimen sudah disimpan ke: komentar_aura_farming_sentimen.csv")
```
